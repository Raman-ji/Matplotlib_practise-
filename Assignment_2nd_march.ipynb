```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
Answer:- A scatter plot is a type of plot used in data visualization that represents individual data points as dots or markers in a two-dimensional coordinate system. It is particularly useful for examining the relationship between two continuous variables and identifying patterns, trends, or clusters in the data.
"""
print("A scatter plot is a type of plot used in data visualization that represents individual data points as dots or markers in a two-dimensional coordinate system. It is particularly useful for examining the relationship between two continuous variables and identifying patterns, trends, or clusters in the data.")
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("this is x data")
plt.ylabel("this is y data")
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""
print("The subplot function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots, each with its own set of axes. This is useful when you want to visualize multiple plots or compare different aspects of your data side by side.The subplot function takes three integer arguments: nrows, ncols, and index. These arguments define the number of rows and columns in the grid of subplots and the index of the specific subplot you want to create.")
import matplotlib.pyplot as plt
import numpy as np
# Line 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2, 2, 1)  # 2 rows, 2 columns, subplot 1
plt.plot(x, y)
plt.title('Line 1')

# Line 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot 2
plt.plot(x, y)
plt.title('Line 2')


# Line 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot 2
plt.plot(x, y)
plt.title('Line 3')


# Line 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, subplot 2
plt.plot(x, y)
plt.title('Line 4')


# Adjust spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
"""Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Answer:-
A bar plot, also known as a bar chart or bar graph, is a type of plot that uses rectangular bars to represent data. It is used to compare and visualize categorical data or discrete values associated with different categories.

Bar plots are commonly used for the following purposes:

1.Comparison: Bar plots are effective for visually comparing the magnitudes or frequencies of different categories. The length or height of each bar represents the value or frequency of the category it represents, allowing for easy comparisons between categories.

2.Distribution: Bar plots can display the distribution of data within different categories. By organizing the bars side by side or in groups, you can observe patterns or variations in the data across categories.

3.Ranking or Sorting: Bar plots can be used to rank or sort categories based on their values. This helps in identifying the highest or lowest values, identifying the most frequent categories, or highlighting the order of importance.

4.Visualizing Relationships: Bar plots can also be used to visualize relationships or associations between different categories or variables. By using different colors or patterns, you can group or differentiate the bars to represent different conditions or variables, providing insights into the relationships between them.

Presentation and Communication: Bar plots are commonly used in presentations, reports, and publications to effectively communicate information to a wide audience. They provide a clear and concise representation of data and facilitate easy understanding and interpretation."""
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
print("Vertical bar plot \n")
plt.bar(company, profit)
plt.show()
print("\n Horizontal bar graph\n")
plt.barh(company, profit)
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Answer:- """
import matplotlib.pyplot as plt
import numpy as np

# Data for the box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axes
fig, ax = plt.subplots()

# Create a box plot
ax.boxplot([box1, box2])

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')

plt.show()

```
